```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
sns.displot(data=titanic, x="fare", kde=False,bins=30)
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
sns.boxplot(data=titanic, x='class', y='age')
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
sns.swarmplot(data=titanic, x='class', y='age',hue='deck',palette='Set2',size=4,legend='auto')
#using palette without hue is deprecated
#palette='Set2' sets the colours of the swarmplot according to the different values for the hue 
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
#above line is used to make sure that the legend is in the upper left corner and not overlapping the plot
#bbox_to_anchor=(1.05, 1) is used to set the location of the legend at co-ordinate (1.05, 1) with respect to the right side of the plot
#the main solution for the required exercise is to use sns.swarmplot(data=titanic, x='class', y='age',palette='Set2')
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
sns.countplot(data=titanic, x='sex')
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
tc=titanic.corr(numeric_only=True)
sns.heatmap(tc,cmap='coolwarm')
```

```python
# CODE HERE
# REPLICATE EXERCISE PLOT IMAGE BELOW
# BE CAREFUL NOT TO OVERWRITE CELL BELOW
# THAT WOULD REMOVE THE EXERCISE PLOT IMAGE!
g=sns.FacetGrid(data=titanic,col='sex')
g.map(plt.hist,'age')
#or use g.map(sns.histplot,'age)
```
